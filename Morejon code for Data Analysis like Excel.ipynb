```python
import pandas as pd
import numpy as np
from scipy import stats

# Load Excel file
df = pd.read_excel("Liliette Data Analysis.xlsx")

# Select column
data = df["NewBalance"].dropna()

# Descriptive statistics
n = len(data)
mean = np.mean(data)
median = np.median(data)
std = np.std(data, ddof=1)          # STDEV.S
var = np.var(data, ddof=1)          # VAR.S
stderr = std / np.sqrt(n)
min_val = np.min(data)
max_val = np.max(data)
range_val = max_val - min_val
skew = stats.skew(data, bias=False)
kurt = stats.kurtosis(data, bias=False)

# 95% Confidence Interval
confidence = 0.95
alpha = 1 - confidence
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
margin_error = t_crit * stderr

LCI = mean - margin_error
UCI = mean + margin_error

# Results DataFrame (rounded to 2 decimals)
results = pd.DataFrame({
    "Statistic": [
        "Count",
        "Mean",
        "Standard Error",
        "Median",
        "Standard Deviation",
        "Sample Variance",
        "Kurtosis",
        "Skewness",
        "Range",
        "Minimum",
        "Maximum",
        "Confidence Level (95%)",
        "Lower Confidence Interval",
        "Upper Confidence Interval"
    ],
    "Value": [
        n,
        round(mean, 2),
        round(stderr, 2),
        round(median, 2),
        round(std, 2),
        round(var, 2),
        round(kurt, 2),
        round(skew, 2),
        round(range_val, 2),
        round(min_val, 2),
        round(max_val, 2),
        round(margin_error, 2),
        round(LCI, 2),
        round(UCI, 2)
    ]
})

results


```
